```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
sns.set(rc = {'figure.figsize': (10, 5)})
```

# 1. Heatmap Basics

- Heapmat **heatmap**
    * Covers basics of heatmap syntax
    * Heatmap takes dataframe as it is and just converts the numbers into a heatmap visualization
    * Learn how to move color bars of heatmap

```python
#load dataset
df = sns.load_dataset('flights')
df.head()
```

```python
#use pivot to transform dataframe
pivot_data = df.pivot("month", "year", "passengers")
pivot_data.head()
```

### 1.1 Basic Syntax

* Heatmap function does not perform any computation, but rather simply converts the numbers into a visualization

```python
#visualize the dataframe into a heatmap
sns.heatmap(pivot_data)
```

```python
#use linewidth parameter to include borders between each cells
sns.heatmap(pivot_data, linewidths = 0.5)
```

```python
#use cmap parameter to change color bars
sns.heatmap(pivot_data, cmap = 'Blues', linewidths = 0.5)
```

### 1.2 Annotations Within Cells

* Use **annot** parameter to include annotations in each cells.
* If numbers are expressed in a scientific format, use **fmt** parameter to force the numbers be expressed in digits.

```python
#annotate cells
sns.heatmap(pivot_data, cmap = 'Blues', linewidths = 0.5, annot = True)
```

```python
#express numbers in a digit format
sns.heatmap(pivot_data, cmap = 'Blues', linewidths = 0.5, annot = True, fmt = 'd')
```

### 1.3 Change Colorbar Location in Visualization

* By default, colorbars are located vertically next to heatmaps. You can change the location of colorbars using subplots.

```python
#create a vertical subplot
fig, (ax, color_ax) = plt.subplots(2)
```

```python
#within one fig, there are two axes
fig
```

```python
#the first axes is named ax as we designed
ax
```

```python
#the second axes is named color_ax as we designed
color_ax
```

```python
#create a heatmap using the same data, but use the ax and cbar_ax parameters to assign in which of the two axes that we created 
#we want to attribute as the visualization axes and as a color axes
heatmap_obj = sns.heatmap(pivot_data, 
                          ax = ax,
                          cbar_ax = color_ax,
                          cbar_kws = {'orientation': 'horizontal'})
```

```python
fig
```

```python
#adjust color bar height and the distance between the two axes
grid_kws = {"height_ratios": (0.85, 0.1), "hspace": 0.4}

fig, (ax, color_ax) = plt.subplots(2, gridspec_kw = grid_kws)

heatmap_obj_v2 = sns.heatmap(pivot_data, 
                          ax = ax,
                          cbar_ax = color_ax,
                          cbar_kws = {'orientation': 'horizontal'})
```
