```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
sns.set(rc = {'figure.figsize': (10, 5)})
```

```python
#functions to be used on this page

def assign_petal_length_class(x):
    """
    Categorizes each row based on its petal length
    """
    if x < 3:
        return 'small'
    elif x < 5:
        return 'medium'
    else:
        return 'large'
    
def assign_petal_length_size(x):
    """
    Assigns a numerical value to each row depending on its petal length
    """
    if x < 3:
        return 3
    elif x < 5:
        return 5
    else:
        return 7
```

# 1. Scatter Graph Basics

- Bar Graphs Using **scatterplot**
    * Covers basics of scatterplot syntax and using the **hue** and **size** parameters to distinguish different categories within a visualization

```python
#load dataset
df = sns.load_dataset('iris')
df.head()
```

```python
#check for null values within dataframe
df.isnull().sum()
```

### 1.1 Basic Scatterplot Syntax

```python
#draw scatterplot
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length')
```

### 1.2 Using the Hue Parameter within Scatterplot

* You can draw a scatterplot based on the X and Y variables, in addition, a third variable on the visualization layer to distinguish each points on the scatter plot
* Using the **Hue** parameter allows you to distinguish each observation in different colors.

```python
#use hue on species variable
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species')
```

### 1.3 Using the Style and Size Parameters

* You can draw a scatterplot based on the X and Y variables, in addition, a third variable on the visualization layer to distinguish groups of points on the scatter plot
* Using the **style** parameter allows you to distinguish groups of observation in different shapes within the visualization.
* Using the **size** parameter allows you to distinguish groups of observation in different sizes within the visualization.
* You can use either the **style** or **size** parameters in addition to the **hue** parameter.

```python
#categorize the petal length using the pre-defined functions
df['petal_length_class'] = df['petal_length'].apply(assign_petal_length_class)
df['petal_length_size'] = df['petal_length'].apply(assign_petal_length_size)

df.head()
```

```python
#use style to distinguish groups of observations in different shapes
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species', style = 'petal_length_class')
```

```python
#use size to distinguish groups of observations in different sizes
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species', size = 'petal_length_size')
```

### 1.4 Using Relplot to Draw FacetGrids

* You can use **Relplot** to draw scatterplot on a FacetGrid.
* Define how each grid should be divided using the **col** parameter.

```python
sns.relplot(data = df, x = 'sepal_width', y = 'sepal_length', col = 'species')
```
