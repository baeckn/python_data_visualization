```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

# 1. Bar Graphs Basics

- Bar Graphs Using **Barplot** vs **Countplot**
    * They are both bar graphs, but the main difference is that **countplot** plots the "count" of the number of recerds by category.
    * On the other hand, **barplot** plots a "value" or "metric" by category

```python
#create a dataframe
movie_title = ['Harry Potter', 'Iron Man', 'Captain America']
audience = [661039, 2039402, 10392]

data = {'movie_title': movie_title, 'audience': audience}
df = pd.DataFrame(data)

df
```

## 1.1 Bar Graph Using Barplot

* Basic bar graph syntax using barplot
* Bar order adjustment
* Label formatting

```python
sns.barplot(data = df, x = 'movie_title', y = 'audience')
```

You can adjust the figure size

```python
#adjust figure size
sns.set(rc={'figure.figsize':(10, 5)})
sns.barplot(data = df, x = 'movie_title', y = 'audience')
```

```python
#barplot order adjustment
sns.barplot(data = df, x = 'movie_title', y = 'audience',
           order = df.sort_values('audience').movie_title)
```

```python
chart = sns.barplot(data = df, x = 'movie_title', y = 'audience',
                    order = df.sort_values('audience').movie_title)

#ylabel formatting
ylabels = ['{:,.0f}'.format(i) + 'K' for i in chart.get_yticks() / 1000]
chart.set_yticklabels(ylabels)
```

```python
#reverse barplot
chart = sns.barplot(data = df, x = 'audience', y = 'movie_title',
                    order = df.sort_values('audience', ascending = False).movie_title)

xlabels = ['{:,.0f}'.format(i) + 'K' for i in chart.get_xticks() / 1000]
chart.set_xticklabels(xlabels)

#change x & y labels and title
plt.xlabel('Cumulative Audience', fontsize = 15)
plt.ylabel('Movie Title', fontsize = 15)
plt.title('Cumulative Audience by Movies', fontsize = 20)
```

## 1.2 Bar Graph Using Countplot

* Basic Syntax
* Hue
* Palette settings

```python
#loading titanic dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
#quickly scan the data
df.describe()
```

```python
#find if there are any null values
df.isnull().sum()
```

```python
#how many passengers were there by gender?
df.groupby(['sex'])['sex'].count().reset_index(name = 'count')
```

* **Countplot** allows you to just insert the dataset and provide by which category you want to count the number of records (in this case male).
* If we were to draw the below using **barplot**, we will have to transform the dataset using groupby like above.

```python
#how many passengers were there by gender?
sns.countplot(data = df, x = 'sex')
```

```python
#how many passengers were there by different classes per gender?
df.groupby(['sex', 'class'])['sex'].count().reset_index(name = 'count')
```

* The **hue** argument is used to provide by which column the visualization use for color encoding.

```python
#using countplot to visualize how many passengers were there by different classes per gender
sns.countplot(data = df, x = 'sex', hue = 'class')
```

* Use the **palette** argument to change the bar colors.
* Click [here](https://seaborn.pydata.org/tutorial/color_palettes.html) for more palette options.

```python
sns.countplot(data = df, x = 'sex', hue = 'class', palette = 'flare')
```
