```python
import matplotlib.pyplot as plt
import pandas as pd
```

# 1. Basics of Matplotlib

* **plt.plot()** generally requires the x and y variables. 
    * If only the y variable is inserted, it assumes an index array from 0 to N - 1 for the x variable
* This section is an overview of how we can change data is displayed

```python
#make data
x = [1, 2, 3]
y = [4, 5, 6]

# plt.figure()
plt.plot(x, y)
```

### 1.1 Change How Data is Displayed

* Linewidth
* Line color
* Marker
* Line style

```python
#control linewidth
plt.plot(x, y, linewidth = 10)
```

```python
#control line color
plt.plot(x, y, color = 'red')
```

```python
#mark point values and control marker size
plt.plot(x, y, marker = 'o', markersize = 10)
```

```python
#change linestyle
plt.plot(x, y, linestyle = ':')
```

```python
#plot using markers instead of lines
plt.plot(x, y, 'o')
```

```python
#use combined argument to denote marker and its color
plt.plot(x, y, 'or')
# plt.plot(x, y, 'o', color='red')
```

### 1. 2 Edit Title and Axis
* Title
* Axis
* Grid
* Annotation
* X and Y limits

```python
#You can insert a title and change its font size
plt.plot(x, y)
plt.title("Title", fontsize = 20)
```

```python
#You can insert x and y labels and change their font sizes
plt.plot(x, y)
plt.xlabel("X Label", fontsize = 20)
plt.ylabel("Y Label", fontsize = 20)
```

```python
#Insert grid for the overall plot
plt.plot(x, y)
plt.grid(True)
```

```python
#You can control in which axis to display grids
plt.plot(x, y)
plt.grid(True, axis = 'x')
#plt.grid(True, axis = 'y')
```

```python
#Set x and y limits
plt.plot(x, y)
plt.xlim([1, 2])
plt.ylim([4, 5])
```

```python
#alternative way to limit axis
plt.plot(x, y)
plt.axis([1, 2, 4, 5])
```

```python
#Insert a 'HERE' text annotation at (2, 5)
plt.plot(x, y)
plt.text(2, 5, 'HERE', fontsize = 20)
```

# 2. Drawing Multiple Data Sources - Single Plot

* What if we want to draw multiple data in a single visualization?
* This section covers:
    * drawing multiple datasource
    * inserting a legend. More about matplotlib legend documentation [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html)

```python
#make data
x1 = [1, 2, 3]
y1 = [1, 2, 3]
x2 = [1, 2, 3]
y2 = [1, 4, 7]

plt.plot(x1, y1, color = 'blue')
plt.plot(x2, y2, color = 'red')
```

```python
#alternative way to insert two distinct data, but this affects readability
plt.plot(x1, y1, 'blue', x2, y2, 'red')
```

```python
#inserting a legend
plt.plot(x1, y1, color = 'blue', label = 'data1')
plt.plot(x2, y2, color = 'red', label = 'data2')
plt.legend()
```

```python
#alternative way to insert a legend
plt.plot(x1, y1, color = 'blue')
plt.plot(x2, y2, color = 'red')
#you can control where to place the legend within the visualization
#as well as its size
plt.legend(['data1', 'data2'], loc = 'upper right', fontsize = 15)
```

# 3. Drawing Multiple Data Sources - Subplot/Subplots

* What if we want to draw multiple data in a different subplots?
* This section covers:
    * subplot
    * subplots

### 3.1 Subplot

```python
#make data
x1 = [1, 2, 3]
y1 = [1, 2, 3]
x2 = [1, 2, 3]
y2 = [1, 100, 200]
```

* subplot() can take three positional arguments
    * number of rows
    * number of columns
    * location of the plot of interest

```python
#set subplot
plt.subplot(1, 2, 1)
plt.plot(x1, y1)
plt.title("Data 1")

plt.subplot(1, 2, 2)
plt.plot(x2, y2)
plt.title("Data 2")
```

```python
#draw subplots
plt.subplot(2, 1, 1)
plt.plot(x1, y1)
plt.title("Data 1")

plt.subplot(2, 1, 2)
plt.plot(x2, y2)
plt.title("Data 2")

#allows no visual overlap between the two charts
plt.tight_layout()
```

### 3.2 Subplots

* Allows you to use a single function to create **a figure** with **several subplots** with only one line of code.
* You can use axes objects to easily access each subplot and build visualizations

```python
#make data
x1 = [1, 2, 3]
y1 = [1, 2, 3]
x2 = [1, 2, 3]
y2 = [1, 50, 200]

#create a figure and axes objects
fig, axe1 = plt.subplots(nrows = 1, ncols = 2)
```

```python
#access each axe objects and build visualizations
axe1[0].plot(x1, y1, color = 'blue')
axe1[1].plot(x2, y2, color = 'red')
fig
```

### 3.3 Double Axis Using Subplots

```python
fig, axe1 = plt.subplots()

axe2 = axe1.twinx()
chart1 = axe1.plot(x1, y1, color = 'red')
chart2 = axe2.plot(x2, y2, color = 'blue')

axe1.set_xlabel('x', fontsize = 15)
axe1.set_ylabel('y1', fontsize = 15)
axe2.set_ylabel('y2', fontsize = 15)

chart = chart1 + chart2
axe1.legend(chart, ['data1', 'data2'])
```
