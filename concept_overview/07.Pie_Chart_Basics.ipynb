```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
sns.set(rc = {'figure.figsize': (10, 5)})
```

# 1. Pie Chart Basics

- Drawing a pie chart using **pie** in matplotlib
    * Covers basics of pie chart syntax
    * Annotating each wedge using labels and colors
    * Learn how to explode each wedge for an advanced visualization

```python
#create data
movie_title = ['Harry Potter', 'Iron Man', 'Captain America']
audience = [661039, 2039402, 110392]

#create a dataframe
data = {'movie_title': movie_title, 'audience': audience}
df = pd.DataFrame(data)

df
```

### 1.1 Basic Syntax

* Similar to a heatmap, the piechart function takes the column value as it is and simply transforms the values into a visualization

```python
#basic syntax
plt.pie(df['audience']);
```

### 1.2 Annotate Wedges with Labels

* Use the **labels** parameter with corresponding labels

```python
#annotate with labels
plt.pie(df['audience'], labels = df['movie_title']);
```

### 1.3 Annotate with Numbers

* The **autopct** parameter automatically converts numberical values into percentages

```python
#add percentage labels
plt.pie(df['audience'], labels = df['movie_title'], autopct = '%0.2f%%');
```

### 1.4 Change Colors

* You can utilize hexacode or text color codes to change wedge colors

```python
#create a color list
colors_list = ['#F08080', '#DCDCDC', '#FFF8DC']

plt.pie(df['audience'], 
        labels = df['movie_title'], 
        autopct = '%0.2f%%',
        colors = colors_list);
```

### 1.5 Explode Wedges

* You can use the **explode** parameter to pass in a list that contains how far from the center each wedge should be displayed.
* By default the first wedge starts from the x-axis and moves counterclockwise. ([link](https://www.w3schools.com/python/matplotlib_pie_charts.asp))

```python
#create a list that defines how far from the center each wedge should be displayed
explode_list = [0, 0.1, 0]

#color code list
colors_list = ['#F08080', '#DCDCDC', '#FFF8DC']

#plot a pie chart
plt.pie(df['audience'], 
        labels = df['movie_title'], 
        autopct = '%0.2f%%',
        colors = colors_list,
        explode = explode_list);
```
