```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
sns.set(rc = {'figure.figsize': (10, 5)})
```

# 1. Line Graph Basics

- Line Graph Using **lineplot**
    * Covers basics of lineplot syntax and using the **hue** and **style** parameters to mark different categories within a visualization

```python
#load dataset
df = sns.load_dataset('flights')
df.head()
```

```python
#check for null values
df.isnull().sum()
```

### 1.1 Basic Syntax

* You can draw a lineplot based on the X and Y variables.

```python
sns.lineplot(data = df, x = 'year', y = 'passengers')
```

### 1.2 Edit the Label Ticks

* You can control how each axis is displayed

```python
#include all years in the x-axis ticks in the visualization
sns.lineplot(data = df, x = 'year', y = 'passengers')
plt.xticks(df['year']);
```

```python
#rotate the x-label ticks 45 degrees
sns.lineplot(data = df, x = 'year', y = 'passengers')
plt.xticks(df['year'], rotation = 45);
```

### 1.3 Use the Hue and Palette Parameters

* Using the **hue** parameter allows you to distinguish groups of observation in different colors in the visualization.
* Using the **style** parameter allows you to distinguish groups of observation in different shapes in the visualization.
* You can customize the colors using the **palette** parameter

```python
#use month as a distinguisher for colors
sns.lineplot(data = df, x = 'year', y = 'passengers', hue = 'month', palette = 'Set2')
```

```python
#use month a distinguisher for both hue and style parameters
sns.lineplot(data = df, x = 'year', y = 'passengers', hue = 'month', palette = 'Set2', style = 'month')
```
