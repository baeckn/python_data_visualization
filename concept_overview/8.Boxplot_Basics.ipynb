```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
sns.set(rc = {'figure.figsize': (10, 5)})
```

# 1. Basics of Boxplot

* Drawing boxplots using **boxplot, swarmplot, violinplot, and stripplot**

```python
#load dataset
df = sns.load_dataset('iris')
df.head()
```

### 1.1 Boxplot Using Boxplot

* Insert dataframe and the plot recognizes the columns as x-axis on the plot.
* You can orient the plot horizontally as well.

```python
#draw boxplot
fig, axe = plt.subplots(figsize = (7, 8))
sns.boxplot(data = df)
```

```python
#orient horizontally
sns.boxplot(data = df, orient = 'h')
```

### 1.2  Boxplot Using Swarmpolt

* **Swarmplot** marks each data point into the visualization

```python
#draw swarmplot
sns.swarmplot(data = df)
```

* You can overlap a **swarmplot** on top of a **boxplot**

```python
#overlap boxplot and swarmplot
plt.subplots(figsize = (10, 10))
sns.boxplot(data = df)
sns.swarmplot(data = df)
```

```python
#change data points colors in a swarmplot
plt.subplots(figsize = (10, 10))
sns.boxplot(data = df)
sns.swarmplot(data = df, color = 'black')
```

### 1.3 Boxplot Using Violinplot

* Violinplot is useful for visualizaing the distribution of data in each given columns

```python
#draw violinplot
sns.violinplot(data = df)
```

### 1.4 Boxplot Using Stripplot

* **Stripplot** is simiar to **swarmplot** except that each data point can overlap on top of each other.
* **Stripplot** prevents data points from being ommitted in visualizations

```python
sns.stripplot(data = df)
```
