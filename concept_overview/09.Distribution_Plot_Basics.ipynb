```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
sns.set(rc = {'figure.figsize': (15, 5)})
```

# Distribution Plot Basics

* There are a few options to draw distribution plots
    * distplot()
    * histplot()
    * displot()
* **distplot()** is to be deprecated so better to use either **histplot or displot**

```python
#load dataset
df = sns.load_dataset('iris')
df.head()
```

### 1.1 Distribution Using Distplot

* An axes-level visualization that takes a series-type data to visualize distribution.
* Change bin size.

```python
#plot
sns.distplot(df['sepal_length'])
```

```python
#change bin size
sns.distplot(df['sepal_length'], bins = 15)
```

### 1.2 Distribution Using Histplot

* Unlike distplot, **histplot** can take either a Series or a dataframe to plot visualization.
* When using a dataframe to visualize, need to specify which column values we want to see distribution of.
* We can differentiate distributions based on a third column values, but in such case we need to utilize the whole dataframe for visualization.
* We can also draw heatmaps using this function.

```python
#plot
sns.histplot(df['sepal_length'])
```

```python
#you can add density curve
sns.histplot(df['sepal_length'], bins = 20, kde = True)
```

```python
#control the y-axis, whether we want to visalize by count or density
sns.histplot(df['sepal_length'], bins = 20, kde = True, stat = 'density')
```

```python
#differentiate by hue.
sns.histplot(data = df, x = 'sepal_length', bins = 20, kde = True, stat = 'density', hue = 'species')
```

```python
#draw heatmap type histplot
sns.histplot(data = df, x = 'sepal_width', y = 'sepal_length', cbar = True)
```

### 1.3 Distribution Using Displot

* While distplot and histplot are axes-level visualizations, **displot** is a figure-level visualization, meaning that we can control visualization settings, such as plot size, directly within the plot command
* Otherwise, its features are very similar to those of **histplot**.

```python
#draw displot. Change plot size and height * aspect = width
sns.displot(data = df, x = 'sepal_length', height = 5, aspect = 3)
```

```python
#using hue similar to histplot
sns.displot(data = df, x = 'sepal_length', bins = 20, kde = True, stat = 'density', hue = 'species')
```

### 1.4 Distribution Using Jointplot

* Jointplot is also a figure-level visualization that enabels joint comparison of distributions of two distinct column values

```python
#draw jointplot
sns.jointplot(data = df, x = 'sepal_width', y = 'sepal_length')
```

```python
#you can change the type of the joint distribution box
sns.jointplot(data = df, x = 'sepal_width', y = 'sepal_length', kind = 'hist', cbar = True)
```

```python
#a figure-level plot so we can directly specify figure size within the command below
sns.jointplot(data = df, x = 'sepal_width', y = 'sepal_length', kind = 'hist', cbar = True, height = 8)
```
